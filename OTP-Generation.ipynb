```python

import random
def otp_generation():
  """ This function will generate random
  6 digits OTP"""
  otp = random.randint(100000,999999)
  return str(otp)

import time

def simulation_otp_to_mail(email,otp):
  """It will simulate that OTP has been
  sent to your e-mail"""

  print(f"Sending otp {otp} to {email}")
  time.sleep(2)
  print("Email 'Sent'!")

def get_otp_from_user():
  """This function asks to enter OTP
  from user"""
  while True:
    user_input = input("Enter OTP: ").strip()
    if len(user_input) == 6 and user_input.isdigit():
      return user_input
    else:
      print("Incorrect, Must be 6 digits")

def otp_verification(actual_otp,entered_otp):
  """ It will secure and match the
  actual OTP to generated OTP """
  otp_match = actual_otp == entered_otp
  return otp_match

def run_otp_verification(max_attempts = 3):
  """ The main function includes all the small function it
  to perform some tasks """
  print("Welcome to the OTP verification System")
  user_email = input('Enter your e-mail id to receive OTP: ').strip()
  if not user_email:
    print("Email address cannot be empty")
    return False

  actual_otp = otp_generation()
  simulation_otp_to_mail(user_email,actual_otp)
  print("Enter OTP for verification")

  attempts_made = 0
  while attempts_made < max_attempts:
    entered_otp = get_otp_from_user()
    if otp_verification(actual_otp,entered_otp):
      print("OTP verification successfull! Access Granted")
      return True
    else:
      attempts_made += 1
      if attempts_made < max_attempts:
        print(f"Incorrect OTP.You have {max_attempts-attempts_made} left.Try again.")
      else:
        print("Incorrect OTP! All attempts Exhausted")

  print("Access Denied! Too many incorrect OTP entered")
  return False

if __name__ == "__main__":
  """if python script runs directly then this __main__ string
  assigned to __name__  variable"""
  if run_otp_verification(max_attempts=3):
    print("Authentication Successfull. You can now access your account")
  else:
    print("Authentication Failde.You can try later or contact support team")
```
